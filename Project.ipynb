```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Customer Churn.csv")
df
```

```python
df['TotalCharges']=df['TotalCharges'].replace(' ','0')
df['TotalCharges']=df['TotalCharges'].astype('float')
```

```python
def conv(val):
    if val==1:
        return 'Yes'
    else:
        return 'No'
df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)
```

```python
df
```

```python
plt.figure(figsize=(5,4))
ax=sns.countplot(data=df,
              x='Churn')
ax.bar_label(ax.containers[0])
plt.title('Count of  Customer by Churn')
plt.show()
```

```python
plt.figure(figsize=(10,5))
gb=df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'],labels=gb.index,autopct="%1.2f%%")
plt.title('Percentage of Churn Customer')
plt.show()
```

```python
# from the given pie chart we can conclude that 26.54% of customers are churn out
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='gender', data=df ,hue='Churn')
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x='SeniorCitizen', data=df )
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='SeniorCitizen', data=df ,hue='Churn')
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python
# Calculate the counts for each combination of SeniorCitizen and Churn
grouped_data = df.groupby(["SeniorCitizen", "Churn"]).size().unstack(fill_value=0)

# Calculate the percentage for each Churn category within each SeniorCitizen group
percent_data = grouped_data.div(grouped_data.sum(axis=1), axis=0) * 100

# Create the stacked bar plot
ax = percent_data.plot(kind='bar', stacked=True, figsize=(5, 3), color=["purple", "Magenta"])

# Set the title and labels
plt.title("Churn by SeniorCitizen")
plt.ylabel("Percentage")
plt.xlabel("SeniorCitizen")

# Add percentage labels on top of the bars
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center', color='white', fontsize=10)

plt.show()
```

```python
#comparatively a greater percentage of people in senior citizen category is churn out
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x='tenure',data=df,bins=72,hue='Churn')
plt.show()
```

```python
# List of columns to create count plots for
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV']

# Set up the matplotlib figure and axes
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))  # 2 rows, 4 columns grid

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop through the columns and create a countplot for each
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i],hue=df["Churn"])
    axes[i].set_title(f'Countplot for {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# The data shows that customers without additional services like OnlineSecurity,
# TechSupport, and DeviceProtection tend to churn more. Fiber optic internet users
# and those without PhoneService or MultipleLines also show higher churn rates. In
# general, having these services seems to reduce customer churn.
```
